Кампанії**

• Мета: Провести A/B тест для оцінки ефективності двох маркетингових кампаній за допомогою статистичного аналізу.

• Набір даних: Використовуйте набір даних для A/B тестування з платформ,

**Завантаження та Попередня Обробка Даних:**

Завантажте набір даних, що містить інформацію про взаємодію
користувачів із двома маркетинговими кампаніями (A і B).

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
control = pd.read_csv('control_group.csv', sep=';')
test = pd.read_csv('test_group.csv', sep=';')
```

Очистіть дані, обробіть відсутні або некоректні значення.

```python
def clean_columns(df):
    df.columns = df.columns.str.strip().str.lower() \
        .str.replace('[usd]', '', regex=False) \
        .str.replace('# of', '') \
        .str.replace(' ', '_') \
        .str.replace('__', '_') \
        .str.strip('_')
    return df

control = clean_columns(control)
test = clean_columns(test)

print("Control DataFrame Columns after cleaning:")
print(control.columns)
print("\nTest DataFrame Columns after cleaning:")
print(test.columns)

print("\nNull values in Control DataFrame:")
print(control.isnull().sum())
print("\nNull values in Test DataFrame:")
print(test.isnull().sum())
```

**Статистичний Аналіз:**

Розрахуйте коефіцієнти конверсії для обох кампаній.

```python
control = control.fillna(control.mean(numeric_only=True))
test = test.fillna(test.mean(numeric_only=True))

c_clicks = control['website_clicks'].sum()
c_impressions = control['impressions'].sum()
c_purchases = control['purchase'].sum()

t_clicks = test['website_clicks'].sum()
t_impressions = test['impressions'].sum()
t_purchases = test['purchase'].sum()

print("Control Group")
print(f"CTR: {(c_clicks / c_impressions) * 100:.2f}%")
print(f"Conversion Rate: {(c_purchases / c_clicks) * 100:.2f}%")

print("\nTest Group")
print(f"CTR: {(t_clicks / t_impressions) * 100:.2f}%")
print(f"Conversion Rate: {(t_purchases / t_clicks) * 100:.2f}%")
```

Проведіть перевірку гіпотези (наприклад, t-тест або тест χ²) для
визначення статистичної значущості різниці в конверсіях.

```python
control_nopurchase = c_clicks - c_purchases

test_nopurchase = t_clicks - t_purchases

contingency_table = np.array([
    [c_purchases, control_nopurchase],
    [t_purchases, test_nopurchase]
])

chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print(f"P-Value (Хі-квадрат): {p_value:.5f}")

if p_value < 0.05:
    print("Результат: Різниця СТАТИСТИЧНО ЗНАЧУЩА (Ми впевнені на 95%)")
else:
    print("Результат: Різниця випадкова (Немає доказів, що одна група краща)")
```

Візуалізуйте результати за допомогою стовпчастих діаграм,
боксплотів і гістограм.

```python
control['Group'] = 'Control'
test['Group'] = 'Test'
df = pd.concat([control, test], ignore_index=True)

plt.figure(figsize=(15, 10))

# Графік 1: Боксплот Покупок
plt.subplot(2, 2, 1)
sns.boxplot(data=df, x='Group', y='purchase', palette='Set2')
plt.title('Де більше стабільних покупок?')

# Графік 2: Боксплот Кліків
plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='Group', y='website_clicks', palette='coolwarm')
plt.title('Де більше кліків?')

# Графік 3: Гістограма розподілу покупок
plt.subplot(2, 2, 3)
sns.histplot(data=df, x='purchase', hue='Group', kde=True, bins=20, alpha=0.5)
plt.title('Як часто купують? (Розподіл)')

# Графік 4: Стовпчики (Сумарні витрати)
plt.subplot(2, 2, 4)
sns.barplot(data=df, x='Group', y='spend', estimator=sum, palette='viridis')
plt.title('Загальні витрати на рекламу')

plt.tight_layout()
plt.show()
```
